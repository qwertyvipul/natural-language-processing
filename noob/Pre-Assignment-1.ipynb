### Frequency Distribution

```python
fdist = FreqDist(text1)
print(fdist)
```

```python
# Get the unique words of the text
vocab = fdist.keys()
print(vocab)
```

```python
# Top 50 words of the text
vocab[:50]
```

```python
# Plot the cumulative frequency graph of the top 50 vocabs
fdist.plot(50, cumulative=True)
```

```python
# Give words which occurs only once
fdist.hapaxes()
```

```python
# Count the number of times the word occured
fdist['monstrous']
```

```python
# Frequency of a given sample
fdist.freq('monstrous')
```

```python
# Total number of samples
fdist.N()
```

```python
# Sample with greatest count
fdist.max()
```
